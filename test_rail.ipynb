```python
import pandas as pd
from service import sncf_routes
from datetime import timedelta,datetime
```

```python
trip1={'origin' :'Aurillac', 'destinations':['Angers','Tours','Nantes']}
trip2={'origin' :'Ussel', 'destinations':['Angers','Tours','Nantes']}
trip3={'origin' :'Tours', 'destinations':['Saint Brieuc']}
trip4={'origin' :'Angers', 'destinations':['Saint Brieuc']}
trip5={'origin' :'Tours', 'destinations':['Angers']}
trip6={'origin' :'Le Mont Dore Gare', 'destinations':['Angers','Tours','Nantes']}
trip7={'origin' :'Nantes', 'destinations':['Hennebont']}

trip_list = [trip1,trip2,trip3,trip4,trip5,trip6,trip7]

dates = []

for delta in range(0,7):
    dt_string = datetime(2017,8,16)+ timedelta(days=delta)
    dates.append(dt_string.isoformat())
```

```python
trip_list
```

```python
df = pd.DataFrame()
for trip in trip_list:
    origin = trip['origin']
    for destination in trip['destinations']:
        for dt in dates:
            df_result = sncf_routes.get_routes(origin,destination,dt)
            df = pd.concat([df,df_result],axis=0)       
```

```python
df.to_csv('train_2017_8_16_to_2017_8_23.csv',index=False,encoding='utf-8')
```

```python
cols_price = [c for c in df.columns if 'price' in c]
```

```python
groupy = df[['trip_id','total_trains','trip_rank','total_duration','trip_departureDate','trip_arrivalDate','origin','destination','trip_name']+cols_price].groupby(['trip_name','trip_departureDate','trip_arrivalDate','total_duration','trip_id']).agg({'min_price':['mean']})

groupy = groupy.reset_index()

groupy.columns = groupy.columns.droplevel(1)

groupy
```
