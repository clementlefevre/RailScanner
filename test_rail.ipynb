```python
import pandas as pd
from service import sncf_routes
from datetime import timedelta,datetime
```

```python
origins =['Aurillac']#,'Angouleme','Brive la Gaillarde']
destinations=['Nantes']
dates = []
for delta in range(0,4):
    dt_string = datetime(2017,8,17)+ timedelta(days=delta)
    dates.append(dt_string.isoformat())
```

```python
dates
```

```python
df = pd.DataFrame()
for origin in origins:
    for destination in destinations:
        for dt in dates:
            df_result = sncf_routes.get_routes(origin,destination,dt)
            df = pd.concat([df,df_result],axis=0)
        
        
```
